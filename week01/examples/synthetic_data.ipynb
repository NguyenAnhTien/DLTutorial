```python
import numpy
import matplotlib.pyplot as plt
```

```python
def generate_synthetic_data(
        n_points=100
):
    """
    @desc:
    - Generate linear data points with noise.
    
    @params:
    - n_points: Number of data points (default is 100).
    - noise_factor: Factor controlling the amount of random noise (default is 0.2).
    
    @returns:
    - x: NumPy array of x values.
    - y: NumPy array of corresponding y values.
    """

    # Generate random x values
    x_data = numpy.linspace(0, 10, n_points)

    # Calculate y values with the given equation and add random noise
    y_points = numpy.sin(2 * numpy.pi * x_data) + numpy.random.normal(size=n_points)

    return x_data, y_points
```

```python
def plot_data(
        x_data: numpy.ndarray,
        y_points: numpy.ndarray,
        fontSize: int=60,
        pointSize: int=300,
        lineWidth: int=20,
        lineColor: str='red',
        figSize: tuple=(50, 40),
        pointColor: int='green'
) -> None:
    """
    @desc:
        - visualize data distribution
    @params:
        - x_data: X-axis values
        - y_points: Y-axis values
    @returns:
        - None
    """
    plt.figure(figsize=figSize)
    plt.scatter(x_data, y_points, label='Data Point', s=pointSize,
                c=pointColor)
    plt.plot(x_data, numpy.sin(2 * numpy.pi * x_data),
             color=lineColor, linestyle='dashed', label='Trending',
             linewidth=lineWidth)
    plt.xlabel('x', fontsize=fontSize)
    plt.ylabel('y', fontsize=fontSize)
    # plt.title('Simulated Data with Noise', fontdict={'fontsize' : fontSize})
    plt.legend(fontsize=fontSize)
    plt.xticks(fontsize=fontSize)
    plt.yticks(fontsize=fontSize)
    plt.show()
```

```python
n_points = 10
```

```python
# Generate random x values
samples = generate_synthetic_data(n_points=n_points)
```

```python
plot_data(samples[0], samples[1], pointSize=1000)
```
